## Machine Learning with LinearRegression 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
```

```python
print(data_clean.head())
```

```python
# assigning target and values
y = data_clean['price'].values
X = data_clean[['sqft', 'bedrooms', 'bathrooms']].values
```

```python
# Measuring accuracy through running it once
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 21)
reg = LinearRegression()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
reg.score(X_test, y_test)
```

## LinearRegression Results 

```python
# Measuring accuracy through cross validation 4 times
# Note: seems to be much lower, lets try to put categorical data in
reg = LinearRegression()
cv_results = cross_val_score(reg, X ,y,cv=4)
print('This ML is ' + str(cv_results.mean()) + ' accurate, given 4 cross validations')
```

```python
print('The R^2 of the model is ' + str(r2_score(y_pred, y_test)))
```

```python
# Measuring accuracy through running it once
predictions = reg.predict(X_test)
plt.scatter(y_test, predictions)
```

```python
sns.distplot(predictions, bins = 40)
```

# Machine Learning with GradientBoostingRegressor
 
Wanted to include region as a variable

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
```

```python
# Transforming region to categorical labels
rle = LabelEncoder()
region_labels = rle.fit_transform(data['region'])
region_mappings = {index: label for index, label in enumerate(rle.classes_)}
data_clean['regionlabel'] = region_labels
region_mappings
```

```python
display(data_clean.head())
```

```python
# assigning target and values
y = data_clean['price'].values
X = data_clean[['sqft', 'bedrooms', 'bathrooms', 'regionlabel']].values
```

```python
# Defining training and testing samples
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 22)

gradientregressor = GradientBoostingRegressor(max_depth = 2, n_estimators = 20, learning_rate = 0.15)
```

```python
# Finding the best combination of GradientBoostingRegressor inputs for highest R^2
LR = {'max_depth': [1,2,3,4],'learning_rate':[0.05,0.1, 0.15,0.2], 'n_estimators': [10,15,20,25]}

tuning = GridSearchCV(estimator = GradientBoostingRegressor(), param_grid = LR, scoring = 'r2')
tuning.fit(X_train, y_train)
tuning.best_params_, tuning.best_score_
```

```python
# Testing the model on just the test set alone
model = gradientregressor.fit(X_train, y_train)
y_pred2 = model.predict(X_test)
model.score(X_test, y_test)
```

## GradientBoostingRegressor Results

```python
# Measuring accuracy through cross validation 5 times
cv_results = cross_val_score(gradientregressor, X ,y ,cv=5)
print('This ML is ' + str(cv_results.mean()) + ' accurate, given 5 cross validations')
```

```python
# Getting the R^2 of the model
print('The R^2 of the model is ' + str(r2_score(y_pred2, y_test)))
```

```python
# Measuring accuracy through running it once
predictions = gradientregressor.predict(X_test)

plt.scatter(y_test, predictions)
```

```python
sns.distplot(predictions, bins = 20)
```
